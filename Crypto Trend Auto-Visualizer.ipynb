# Automating and Visualizing Cryptocurrency Trends with CoinMarketCap API

## Project Overview

This project automates the analysis of cryptocurrency data from the CoinMarketCap API. The data scraping process was executed 333 times, with a 60-second interval between each run. Data cleaning was performed to ensure accuracy and consistency.

## Visualizations Created

*   **Percent Change Over Time:** A categorical plot showing how cryptocurrency values changed across different time frames (1 hour, 24 hours, 7 days, etc.).
*   **Price vs. Time:** A line plot visualizing price movements over time, enabling better trend analysis.

These visualizations provide insights into market behavior, allowing users to interpret value fluctuations and make data-driven decisions in the cryptocurrency market.

## Libraries Used

The following Python libraries were utilized in this project:

*   `requests`: For making HTTP requests to the CoinMarketCap API.
*   `json`: To parse and manage the data fetched from the API.
*   `pandas`: For data manipulation and cleaning.
*   `seaborn`: To create insightful and aesthetic visualizations.
*   `matplotlib.pyplot`: For additional plotting and customization.
*   `os`: To manage file paths and directories.
*   `time` & `sleep`: For managing intervals during automated data fetching.

## How It Works

### Data Fetching

Data was retrieved programmatically using the CoinMarketCap API. Requests were made in a loop with a 60-second pause between each run to avoid API rate limits.

### Data Cleaning

The raw data was cleaned using Pandas to remove inconsistencies and prepare it for analysis.

### Visualization

Seaborn was used to create the visualizations. Matplotlib was utilized for further customization.

```python
# Libraries 
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from time import time
from time import sleep

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1', # Index where you want it to start from
      'limit':'10', # Number of rows you want to extract
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'API_KEY', # enter your coinmarketcap's API Key.
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      # print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
    
    
    
    df = pd.json_normalize(data['data'])
    df['timestamp'] = pd.to_datetime('now')

    if os.path.isfile('api.csv'):
        df.to_csv('api.csv', index= False, mode = 'a', header= None)
    else:
        df.to_csv('api.csv',index = False, header= 'column names')
```

```python


for i in range(333):
    api_runner()
    print('Data Fetched Successfully.')
    sleep(60)
    
exit()
```

```python
df3 = df2.groupby('name',sort = False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d',
            'quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
df3
```

```python
df6= df5.reset_index()
```

```python
df7= df6.rename(columns={'level_1' : 'percent change'} )
df7
```

```python
df7['percent change'] = df7['percent change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d',
            'quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d'],
                             ['1H','24H','7D','30D','60D','90D'])
df7
```

```python
sns.catplot(x='percent change',y='values', hue='name',data= df7, kind='point')
plt.title('Percent Change with respect to time')
plt.show()
```
